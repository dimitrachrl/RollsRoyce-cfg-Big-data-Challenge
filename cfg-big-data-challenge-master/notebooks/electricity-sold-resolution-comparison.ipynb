```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import matplotlib.ticker as ticker

# Read in annual data.
elec_sold_a = pd.read_excel('../data/electricity_clean.xlsx', sheet_name='6.0')

# Read in quarter data.
elec_sold_q = pd.read_excel('../data/electricity_clean.xlsx', sheet_name='6.1')
elec_sold_q['year_quarter'] = elec_sold_q['year'].astype(str) + '-' + elec_sold_q['quarter'].astype(str)

# Read in month data.
elec_sold_m = pd.read_excel('../data/electricity_clean.xlsx', sheet_name='6.2')
elec_sold_m['date'] = elec_sold_m['year'].astype(str) + '-' + elec_sold_m['month'].astype(str)
elec_sold_m['date'] = pd.to_datetime(elec_sold_m['date'],
                                       format='%Y-%B')
```

# Annually 

```python
start_date = 1995
end_date = 2019

elec_sold_a_plot = elec_sold_a.loc[(elec_sold_a['year'] >= start_date) & 
                                   (elec_sold_a['year'] <= end_date)]

fig, ax = plt.subplots(figsize=(15, 7))

plt.plot(elec_sold_a_plot['year'], 
         elec_sold_a_plot['electricity_sold_total'], 
         color='darkgoldenrod')

ax.set_title('Total Electricity Sold Annually by the Major Power Producers')
ax.set_xlabel('Year')
ax.set_ylabel('Electricity sold (TWh)')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

# Quarterly

```python
start_date = 1995
end_date = 2019

elec_sold_q_plot = elec_sold_q.loc[(elec_sold_q['year'] >= start_date) & 
                                   (elec_sold_q['year'] <= end_date)]

fig, ax = plt.subplots(figsize=(15, 7))
plt.xticks(rotation=45)
plt.plot(elec_sold_q_plot['year_quarter'],
         elec_sold_q_plot['electricity_sold_total'],
         color='darkgoldenrod',
         label='Electricity sold')

ax.set_title('Total Electricity Sold each Quarter by the Major Power Producers')
ax.set_xlabel('Year')
ax.set_ylabel('Electricity sold (TWh)')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

tick_spacing = 4
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
```

# Monthly

```python
start_date = '1995-01-01'
end_date = '2019-12-01'

elec_sold_m_plot = elec_sold_m.loc[(elec_sold_m['date'] >= start_date) & 
                                   (elec_sold_m['date'] <= end_date)]

fig, ax = plt.subplots(figsize=(15, 7))

plt.plot(elec_sold_m_plot['date'],
         elec_sold_m_plot['electricity_sold_total'],
         color='darkgoldenrod',
         label='Electricity sold')

ax.set_title('Total Electricity Sold Monthly by the Major Power Producers')
ax.set_xlabel('Year')
ax.set_ylabel('Electricity sold (TWh)')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

date_form = DateFormatter("%Y")
ax.xaxis.set_major_formatter(date_form)
```
