```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime



# Import the data
df = pd.read_excel('Data\supply_prediction.xls')


df.head()
```

```python
# Set date column as index
df.set_index('date',inplace=True)

# Remove no longer required columns
df = df.drop(columns=['electricity_available First Difference', 'Seasonal First Difference'])

df.head()
```

```python
df.plot(xlabel = "",
        ylabel = "Electricity available (TWh)",
        title = "Predicted vs actual monthly electricity supply in the UK: Historical and Predicted Data",
        figsize=(12, 8))
```
